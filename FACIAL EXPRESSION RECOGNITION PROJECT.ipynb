# Dataset Splitting Script for Facial Expression Recognition

```python
import os
import shutil
import warnings
from sklearn.model_selection import train_test_split

warnings.filterwarnings('ignore')

def prepare_splits(input_dir, output_dir, train_ratio=0.7, val_ratio=0.15, test_ratio=0.15):
    """
    Splits the dataset into train, validation, and test folders.

    Args:
        input_dir (str): Path to the folder containing the raw dataset.
        output_dir (str): Path to save the split dataset folders.
        train_ratio (float): Proportion of training data.
        val_ratio (float): Proportion of validation data.
        test_ratio (float): Proportion of test data.
    """
    assert train_ratio + val_ratio + test_ratio == 1, "Ratios must sum to 1."

    for class_name in os.listdir(input_dir):
        class_path = os.path.join(input_dir, class_name)
        if os.path.isdir(class_path):  # Ensure it's a folder
            file_names = os.listdir(class_path)

            # Split data
            train_files, temp_files = train_test_split(file_names, test_size=1-train_ratio, random_state=42)
            val_files, test_files = train_test_split(temp_files, test_size=test_ratio/(test_ratio + val_ratio), random_state=42)

            # Copy files into train, val, and test folders
            for split, files in zip(["train", "validation", "test"], [train_files, val_files, test_files]):
                split_dir = os.path.join(output_dir, split, class_name)
                os.makedirs(split_dir, exist_ok=True)
                
                for file in files:
                    shutil.copy(os.path.join(class_path, file), os.path.join(split_dir, file))

    print("Dataset splitting complete!")

# Define the input and output folder paths
input_dataset_dir = r"C:\Users\hp\OneDrive\Documents\Desktop\Project\images"  
output_dataset_dir = r"C:\Users\hp\OneDrive\Documents\Desktop\Project\split_dataset" 

# Call the function
prepare_splits(input_dataset_dir, output_dataset_dir)
```

# Importing required libraries

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import load_img, img_to_array
import cv2
from mtcnn import MTCNN
```

# Dataset preparation with Image Augmentation

```python
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

validation_datagen = ImageDataGenerator(rescale=1.0 / 255)
```

# Loading and Preprocessing Image Data

```python
# Loading Training Dataset
train_dataset = train_datagen.flow_from_directory(
    r"C:\Users\hp\OneDrive\Documents\Desktop\Project\train",  # Use raw string for paths
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

# Loading Validation Dataset
validation_dataset = validation_datagen.flow_from_directory(
    r"C:\Users\hp\OneDrive\Documents\Desktop\Project\validation",  
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
```

# Model Architecture with Transfer Learning

```python
base_model = MobileNetV2(weights="imagenet", include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Freeze the base model layers
```

# Building the Facial Expression Classification Model

```python
from tensorflow.keras.layers import GlobalAveragePooling2D

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),  # Replaces Flatten to retain spatial features
    Dense(256, activation='relu'),
    BatchNormalization(),  # Apply before activation for stability
    Dropout(0.5),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(train_dataset.num_classes, activation='softmax')  # Ensure train_dataset is defined
])
```

# Compiling the Model for Training

```python
model.compile(
    optimizer=Adam(learning_rate=1e-3),  # Equivalent to 0.001 but more readable
    loss='categorical_crossentropy',
    metrics=['accuracy', 'Precision', 'Recall']  # Adding Precision & Recall for better evaluation
)
```

# Implementing Callbacks for Training Optimization

```python
callbacks = [
    EarlyStopping(monitor='val_loss', mode='min', patience=5, restore_best_weights=True),  
    ReduceLROnPlateau(monitor='val_loss', mode='min', factor=0.2, patience=3, min_lr=1e-6)
]
```

# Training the Model with Augmented Dataset

```python
history = model.fit(
    train_dataset,
    steps_per_epoch=len(train_dataset),  # Uses full dataset per epoch
    validation_data=validation_dataset,
    validation_steps=len(validation_dataset),
    epochs=30,
    callbacks=callbacks
)
```

# Face Detection and Real-Time Emotion Prediction

```python
# Face Detection and Real-Time Prediction
def detect_and_predict_faces(image_path, model, target_size=(224, 224)):
    detector = MTCNN()
    img = cv2.imread(image_path)
    
    if img is None:
        print("Error: Unable to read image.")
        return

    faces = detector.detect_faces(img)

    if not faces:
        print("No face detected.")
        return

    class_labels = {0: "Angry", 1: "Happy", 2: "Sad"}

    for face in faces:
        x, y, width, height = face['box']
        
        # Ensure cropping is within image boundaries
        x, y = max(0, x), max(0, y)
        cropped_face = img[y:y+height, x:x+width]
        
        cropped_face = cv2.resize(cropped_face, target_size)
        cropped_face = cropped_face / 255.0  # Normalize
        cropped_face = np.expand_dims(cropped_face, axis=0)  # Reshape for model input

        # Make prediction
        prediction = model.predict(cropped_face)
        class_index = np.argmax(prediction)
        
        print(f"Predicted Class: {class_labels.get(class_index, 'Unknown')} (Confidence: {np.max(prediction):.2f})")
```

# Real-Time Facial Expression Recognition Using Webcam

```python
import cv2
import numpy as np
from mtcnn import MTCNN

# Load face detector and open webcam
detector = MTCNN()
cap = cv2.VideoCapture(0)  # Open webcam

class_labels = {0: "Angry", 1: "Happy", 2: "Sad"}

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Detect faces in the frame
    faces = detector.detect_faces(frame)

    for face in faces:
        x, y, width, height = face['box']

        # Ensure cropping remains within image boundaries
        x, y = max(0, x), max(0, y)
        cropped_face = frame[y:y+height, x:x+width]

        # Preprocess face for model prediction
        cropped_face = cv2.resize(cropped_face, (224, 224)) / 255.0
        cropped_face = np.expand_dims(cropped_face, axis=0)

        # Make prediction
        prediction = model.predict(cropped_face)
        class_index = np.argmax(prediction)
        confidence = np.max(prediction)

        # Format label text
        label = f"{class_labels.get(class_index, 'Unknown')}: {confidence:.2f}"

        # Draw bounding box & label on the frame
        cv2.rectangle(frame, (x, y), (x+width, y+height), (0, 255, 0), 3)  # Green bounding box
        cv2.putText(frame, label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)  # White text

    # Show the frame
    cv2.imshow("Emotion Recognition", frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
